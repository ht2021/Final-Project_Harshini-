```python
raw_data_IV3 = pd.read_csv("C:/Users/VIOLIN/Desktop/Final-Project_Harshini-/Project Part -2/Data part2/urbanization.csv")
##Printing data
raw_data_IV3.head()
```

The urbanization percentages of 1960 and 2019 are compared to see the relative growth in urbanization over the years for each country!

```python
IV3 = raw_data_IV3[["Country Name","1960","2019"]]

IV3.columns = ['country','1960 urban population%','2019 urban population%']
IV3['change of urbanization % since 1960'] = IV3['2019 urban population%'] - IV3['1960 urban population%']

IV3.head()
```

```python
#checking how much invalid data exists
    # to determine if data is valid
IV3.isnull().sum()
```

Thankfully, there are not many missing values even in the 1960 year!!

```python
## Let us fill the missing values with -999

IV3.replace(to_replace = np.nan, value = -999) 
```
