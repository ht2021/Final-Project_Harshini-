## Sustainability parameter 6 - Infrastructure Facilities

```python
## Importing required packages
import pandas as pd 
import numpy as np 
```

```python
raw_data_infra = pd.read_excel("C:/Users/VIOLIN/Desktop/Final-Project_Harshini-/Data/Global CPi Database.xlsx",sheet_name ="Table 1.")
##Printing data
raw_data_infra.head()
```

```python
## Pre processing the data
raw_data_infra.rename({'Unnamed: 1': 'country', 'Unnamed: 9':'infra_index'}, axis=1, inplace=True)
raw_data_infra.head()
```

```python
data = raw_data_infra[['country','infra_index']]
data.head()
```

```python
infra_data = data.drop([0])
infra_data.head()
```

```python
 #checking how much invalid data exists
    # to determine if data is valid
infra_data.isnull().sum()
```

```python
infra_data.dtypes
```

```python
infra_data['infra_index'] = pd.to_numeric(infra_data['infra_index'],errors='coerce')
```

```python
infra_data.dtypes
```

```python
count = infra_data.groupby('country').infra_index.count()

count = count.reset_index()
print(count) 
```

```python
count.rename({ 'infra_index':'Cities in each country'}, axis=1, inplace=True)
count.head()
```

```python
sum = infra_data.groupby('country').infra_index.sum()

sum = sum.reset_index()
print(sum) 
```

```python
sum.rename({ 'infra_index':'Sum of values of infra indexes of cities'}, axis=1, inplace=True)
sum.head()
```

```python
## Let us merge these datasets

output = pd.merge(count,sum,on ='country',how = "inner")

output.head()
```

```python
output= output[output['Cities in each country'] != 0]
output = output.reset_index()
output.head()
```

```python
SP6_data = output[['country','Cities in each country','Sum of values of infra indexes of cities'] ]
SP6_data['Avg infra index of country'] = SP6_data['Sum of values of infra indexes of cities']/SP6_data['Cities in each country']
SP6_data.head()
```

```python
SP6 = SP6_data[['country','Avg infra index of country'] ]
SP6.head()
```
