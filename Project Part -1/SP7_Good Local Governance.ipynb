```python
## Importing required packages
import pandas as pd 
import numpy as np 
```

```python
raw_data_GG = pd.read_excel("C:/Users/VIOLIN/Desktop/Final-Project_Harshini-/Data/Global CPi Database.xlsx",sheet_name ="Table 4")
##Printing data
raw_data_GG.head(10)
```

```python
## Pre processing the data
raw_data_GG.rename({'Unnamed: 0': 'country', 'URBAN GOVERNANCE AND LEGISLATION INDEX (GI)':'UGI'}, axis=1, inplace=True)
raw_data_GG.head()
```

```python
data = raw_data_GG[['country','UGI']]
data.head()
```

```python
GG_data = data.drop([0,1,2])
GG_data.head()
```

```python
GG_data.dtypes
```

```python
GG_data['UGI'] = pd.to_numeric(GG_data['UGI'],errors='coerce')
GG_data.dtypes
```

```python
 #checking how much invalid data exists
    # to determine if data is valid
GG_data.isnull().sum()
```

```python
count = GG_data.groupby('country').UGI.count()

count = count.reset_index()
print(count) 
```

```python
count.rename({ 'UGI':'Cities in each country'}, axis=1, inplace=True)
count.head()
```

```python
sum = GG_data.groupby('country').UGI.sum()

sum = sum.reset_index()
print(sum) 
```

```python
sum.rename({ 'UGI':'Sum of values of Urban Governance Index'}, axis=1, inplace=True)
sum.head()
```

```python
## Let us merge these datasets

output = pd.merge(count,sum,on ='country',how = "inner")

output.head()
```

```python
output= output[output['Cities in each country'] != 0]
output = output.reset_index()
output.head()
```

```python
SP7_data = output[['country','Cities in each country','Sum of values of Urban Governance Index'] ]
SP7_data['Avg Urban Gov index of country'] = SP7_data['Sum of values of Urban Governance Index']/SP7_data['Cities in each country']
SP7_data.head()
```

```python
SP7 = SP7_data[['country','Avg Urban Gov index of country'] ]
SP7.head()
```
